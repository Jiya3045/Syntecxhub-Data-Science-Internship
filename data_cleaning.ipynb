```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df["venue"] = pd.to_numeric(df["venue"], errors="coerce")
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]

removed_duplicates = before - after
```

```python
df.to_csv("cleaned_data/cleaned_data.csv", index=False)
```

```python
with open("cleaning_log.txt", "w") as f:
    f.write("DATA CLEANING LOG\n")
    f.write("-----------------\n")
    f.write(f"Total rows after cleaning: {df.shape[0]}\n")
    f.write(f"Duplicates removed: {removed_duplicates}\n")
    f.write("Missing values handled using mean/mode\n")
    f.write("Column names standardized\n")
    f.write("Incorrect data types fixed\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
os.makedirs("images", exist_ok=True)


```

```python
df = pd.read_csv("C:/Users/shree/Desktop/sales_data.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
plt.figure()
plt.plot(df['Date'], df['Sales'])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/line_sales.png")
plt.show()

```

```python
monthly_sales = df.resample('M', on='Date')['Sales'].sum()

plt.figure()
monthly_sales.plot()
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales")
plt.tight_layout()
plt.savefig("images/monthly_sales.png")
plt.show()

```

```python

category_sales = df.groupby('Category')['Sales'].sum()

plt.figure()
category_sales.plot(kind='bar')
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Sales by Category")
plt.tight_layout()
plt.savefig("images/bar_category.png")
plt.show()
```

```python
plt.figure()
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.title("Category Share")
plt.tight_layout()
plt.savefig("images/pie_category.png")
plt.show()

```
